```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#pca
df=pd.read_csv(r"C:\Users\hanal\Downloads\8836201-6f9306ad21398ea43cba4f7d537619d0e07d5ae3.csv")
ddf=df.drop('variety',axis=1)
sc=StandardScaler()
sc_df=pd.DataFrame(sc.fit_transform(ddf))
sc_df

```

```python
sc_df.corr()
```

```python
print("size : ",np.shape(ddf))
```

```python
sns.heatmap(sc_df.corr())
```

```python
pca=PCA(n_components=2)
d_pca=pca.fit_transform(ddf)
print("size : ",np.shape(d_pca))
d_pca=pd.DataFrame(d_pca,columns=['PC1','PC2'])
d_pca.head()
```

```python
d_pca.corr()
```

```python
import seaborn as sns
sns.distplot(df["petal.width"],hist=False)
```

```python
print("size : ",np.shape(d_pca))
print("variance after pca:",pca.explained_variance_ratio_.sum()*100)
```

```python

d_pca['variety']=df['variety']
sns.scatterplot(data=d_pca,x='PC1',y='PC2',hue='variety')
```

```python
xx=d_pca.iloc[:,:-1]
yy=d_pca.iloc[:,-1]

xx_train, xx_test, yy_train, yy_test = train_test_split(xx, yy, test_size=0.2, random_state=42)
```

```python

svm_l2 = SVC(kernel='linear', C=1,random_state=5)
svm_l2.fit(xx_train, yy_train)
pl2=svm_l2.predict(xx_test)
print(classification_report(yy_test,pl2))
al2=accuracy_score(yy_test,pl2)
print("accuracy:",al2)
```

```python
svm_nl2 = SVC(kernel='rbf', C=1,random_state=5)
svm_nl2.fit(xx_train, yy_train)
pnl2=svm_nl2.predict(xx_test)
print(classification_report(yy_test,pnl2))
bnl2=accuracy_score(yy_test,pnl2)
print("accuracy:",bnl2)
```

```python
ac={ 'before pca':[al1,anl1],'after pca':[al2,bnl2]}
re=pd.DataFrame(ac,index=['linear',' non-linear'])

print('ACCURACY:')   
re
```
