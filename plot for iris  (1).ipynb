```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy as sy


df=pd.read_csv(r"C:\Users\hanal\Downloads\8836201-6f9306ad21398ea43cba4f7d537619d0e07d5ae3.csv")
print("mean of sepal width :",round(np.mean(df['sepal.width']),2))
print("mean of sepal length :",round(np.mean(df['sepal.length']),2))
print("mean of petal width :",round(np.mean(df['petal.width']),2))
print("mean of petal length :",round(np.mean(df['petal.length']),2))

print("\n\nfor petal length :\n    ",sy.stats.mode(df['petal.length']))
print("for petal width :\n    ",sy.stats.mode(df['petal.width']))
print("for sepal length :\n    ",sy.stats.mode(df['sepal.length']))
print("for sepal width :\n    ",sy.stats.mode(df['sepal.width']))

print("\nmedian of petal length :",np.median(df['petal.length']))
print("median of petal width :",np.median(df['petal.width']))
print("median of sepal length :",np.median(df['sepal.length']))
print("median of sepal width :",np.median(df['sepal.width']))

print("\nstandard deviation of petal length :",round(np.std(df['petal.length']),2))
print("standard deviation of petal width :",round(np.std(df['petal.width']),2))
print("standard deviation of sepal length :",round(np.std(df['sepal.length']),2))
print("standard deviation of sepal width :",round(np.std(df['sepal.width']),2))
```
