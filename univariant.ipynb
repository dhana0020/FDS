```python
# 3

import pandas as pd
df=pd.read_csv(r"C:\Users\hanal\Downloads\diabetes_data_all_patient (1).csv")
print("STANDARD DEVIATION:")
print("patient id : ",df['patient_id'].std())
print("code : ",df['code'].std())

df['value']=pd.to_numeric(df['value'],errors='coerce')#ignore non numeric values
print("value : ",df['value'].std())

df['date'] = pd.to_datetime(df['date'], format="%m-%d-%Y", errors='coerce')
print("date:", df['date'].std())

df['time'] = pd.to_datetime(df['time'], format="%H:%M", errors='coerce')
df['time'] = pd.DatetimeIndex(df['time'])
df['time_m'] = df['time'].dt.hour * 60 + df['time'].dt.minute
time= df['time_m'][df['time_m'].notna()].std()
hours = int(time // 60)
minutes = int(time % 60)
print("time:", f"{hours}:{minutes:02d}")
```

```python
# 4

import pandas as pd
df=pd.read_csv(r"C:\Users\hanal\Downloads\diabetes_data_all_patient (1).csv")
print("SKEWNESS:")

print("patient id : ",df['patient_id'].skew())
print("code : ",df['code'].skew())

df['value']=pd.to_numeric(df['value'],errors='coerce')#ignore non numeric values
print("value : ",df['value'].skew())

df['time'] = pd.to_datetime(df['time'], format="%H:%M", errors='coerce')
df['time'] = pd.DatetimeIndex(df['time'])

time_s = df['time'].dt.hour * 60 + df['time'].dt.minute
skewness = time_s.skew()
print("time:", skewness)

df['date'] = pd.to_datetime(df['date'], format="%m-%d-%Y", errors='coerce')
min_date = df['date'].min()
df['date_num'] = (df['date'] - min_date).dt.days
print("date:", df['date_num'].skew())
```

```python
# 5

import pandas as pd
df=pd.read_csv(r"C:\Users\hanal\Downloads\diabetes_data_all_patient (1).csv")
print("VARIANCE:")
print("patient id : ",df['patient_id'].var())
print("code : ",df['code'].var())

df['value']=pd.to_numeric(df['value'],errors='coerce')#ignore non numeric values
print("value : ",df['value'].var())

df['date'] = pd.to_datetime(df['date'], format='%m-%d-%Y',errors='coerce') 
min_date = df['date'].min() 
df['date_num'] = (df['date'] - min_date).dt.days 
print("date:",df['date_num'].var())


df['time'] = pd.to_datetime(df['time'], format="%H:%M", errors='coerce')
df['time'] = pd.DatetimeIndex(df['time'])
df['time_m'] = df['time'].dt.hour * 60 + df['time'].dt.minute
var = df['time_m'][df['time_m'].notna()].var()
min = int(var // 60)
sec = int(var % 60)
print(f"time: {min}:{sec:02d}")
```

```python
#6

import pandas as pd
import scipy as sy
df=pd.read_csv(r"C:\Users\hanal\Downloads\diabetes_data_all_patient (1).csv")
print("MODE:")
print("patient id: ",sy.stats.mode(df['patient_id']))

print("code : ",sy.stats.mode(df['code']))

df['value']=pd.to_numeric(df['value'],errors='coerce')#ignore non numeric values
print("value : ",sy.stats.mode(df['value']))


df['date'] = pd.to_datetime(df['date'], format="%m-%d-%Y", errors='coerce')
df.dropna(subset=['date'], inplace=True)  # remove rows with missing values in 'date' column
df['date_unix'] = df['date'].apply(lambda x: x.timestamp())
mode_value, count = sy.stats.mode(df['date_unix'])
mode_date = pd.to_datetime(mode_value, unit='s')
print(f"date: {mode_date}, Count: {count}")

df['time'] = pd.to_datetime(df['time'], format="%H:%M", errors='coerce')
df['time'] = pd.DatetimeIndex(df['time'])
df['time_m'] = df['time'].dt.hour * 60 + df['time'].dt.minute
mode_result = sy.stats.mode(df['time_m'][df['time_m'].notna()])
mode_value = mode_result.mode
count = mode_result.count
hours = int(mode_value // 60)
minutes = int(mode_value % 60)
print(f"Mode: {hours:02d}:{minutes:02d}, Count: {count}")
```

```python
#7

import pandas as pd
df=pd.read_csv(r"C:\Users\hanal\Downloads\diabetes_data_all_patient (1).csv")
print("KURTOSIS:")
print("patient id:",df['patient_id'].kurtosis())

print("code : ",df['code'].kurtosis())

df['value']=pd.to_numeric(df['value'],errors='coerce')#ignore non numeric values
print("value : ",df['value'].kurtosis())


df['time'] = pd.to_datetime(df['time'], format="%H:%M", errors='coerce')
df['time'] = pd.DatetimeIndex(df['time'])
time_s = df['time'].dt.hour * 60 + df['time'].dt.minute
print("time:",time_s.kurtosis())


df['date'] = pd.to_datetime(df['date'], format="%m-%d-%Y", errors='coerce')
min_date = df['date'].min()
df['date_num'] = (df['date'] - min_date).dt.days
print("date:", df['date_num'].kurtosis())
```
