```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv(r"C:\Users\hanal\Downloads\dataset.csv")
```

```python

df.head()
```

```python
print("size : ",np.shape(df))
df.corr()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
svm_l1 = SVC(kernel='linear')
svm_l1.fit(x_train, y_train)
pl1=svm_l1.predict(x_test)
print(classification_report(y_test,pl1))
al1=accuracy_score(y_test,pl1)
print("accuracy:",al1)
```

```python

svm_nl1 = SVC(kernel='rbf')
svm_nl1.fit(x_train, y_train)
pnl1=svm_nl1.predict(x_test)
print(classification_report(y_test,pnl1))
anl1=accuracy_score(y_test,pnl1)
print("accuracy:",anl1)
```

```python
import statsmodels.formula.api as a
import numpy as np
b=a.ols(formula='target~trestbps+chol',data=df)
c=b.fit()
c.params
```

```python
xx,yy=np.meshgrid(np.linspace(df.trestbps.min(),df.trestbps.max()),np.linspace(df.chol.min(),df.chol.max()))
xxx=pd.DataFrame({'trestbps':xx.ravel(),'chol':yy.ravel()})
f=c.predict(exog=xxx)
f=np.array(f)
z=plt.figure()
p=z.add_subplot(111,projection='3d')
p.scatter(df['trestbps'],df['chol'],df['target'])
p.plot_surface(xx,yy,f.reshape(xx.shape),color='red')
plt.xlabel('trestbps')

```

```python
from sklearn.model_selection import train_test_split as tt
from sklearn.linear_model import LinearRegression as lr
df=pd.read_csv(r"C:\Users\hanal\Downloads\dataset.csv")
xx=df.iloc[:,:-1]
yy=df.iloc[:,-1]
xx_train,xx_test,yy_train,yy_test=tt(xx,yy,test_size=0.2,random_state=42)
lin=lr()
lin.fit(xx_train,yy_train)
y_p=lin.predict(xx_train)
plt.scatter(yy_train, y_p)
plt.xlabel('Actual')
plt.ylabel('Predicted')
s, i= np.polyfit(yy_train, y_p, 1)  
r = s * np.array(yy_train) + i
plt.plot(yy_train, r, color='red', linestyle='--')

```
