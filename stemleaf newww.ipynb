```python
import pandas as pd
df=pd.read_csv(r"C:\Users\hanal\Downloads\dataset.csv")
df
```

```python
import matplotlib.pyplot as plt

def s_l(d):
    stem = {}
    for i in d:
        s = i // 10
        l = i % 10
        if s not in stem:
            stem[s] = []
        stem[s].append(l)

    for i, j in sorted(stem.items()):
        print(f" {i}  | {', '.join(map(str, sorted(j)))}")

    # Create a stem plot
    stem_x = []
    stem_y = []
    for i, j in sorted(stem.items()):
        for l in j:
            stem_x.append(i)
            stem_y.append(l)

    plt.stem(stem_y, stem_x)
    plt.xlabel('leaf')
    plt.ylabel('stem')
   
    plt.show()

```

```python
chol = df['chol'].tolist()
s_l(chol)
```
